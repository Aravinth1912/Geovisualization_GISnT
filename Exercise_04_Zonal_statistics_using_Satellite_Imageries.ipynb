```python
modis = ee.ImageCollection("MODIS/006/MOD44B")
visualization = {
  'bands': ['Percent_Tree_Cover'],
  'min': 0,
  'max': 100,
  'palette': ['bbe029', '0a9501', '074b03']
};
m.addLayer(modis, visualization, 'Percent Tree Cover')
```

```python
m
```

```python
districts = ee.FeatureCollection('projects/ee-aravinthraja/assets/India_districts')
m.addLayer(districts,{},'districts')
```

```python
India_vegetation = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2019-01-01","2019-12-31").select('Percent_Tree_Cover').mean().clip(districts)
m.addLayer(India_vegetation,visualization,'India_vegetation')
m
```

```python
out_dir = os.path.join(os.path.expanduser("~"), "Downloads")
out_NDVI_stats = os.path.join(out_dir, "NDVI_SAMPLE.csv")

if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# Allowed output formats: csv, shp, json, kml, kmz
# Allowed statistics type: MEAN, MAXIMUM, MINIMUM, MEDIAN, STD, MIN_MAX, VARIANCE, SUM
geemap.zonal_stats(India_vegetation, districts, out_NDVI_stats, stat_type="MEAN", scale=250)
```
